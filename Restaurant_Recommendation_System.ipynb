```python
def handle_rate(rate):
    rate=rate.strip()
    match=re.match(r"([0-9.]+)\s*",rate)
    if match:
        return float(match.group(1))
    else:
        return np.nan
    
df['rate']=df['rate'].apply(handle_rate)
df['rate']
```

```python
import string
punct_to_remove=string.punctuation
def remove_punctuation(txt):
    return txt.translate(str.maketrans('','',punct_to_remove))
df['reviews_list']=df['reviews_list'].apply(lambda txt:remove_punctuation(txt))
df[['reviews_list']].sample(3)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
STOPWORDS=set(stopwords.words('english'))
def remove_stopwords(txt):
    return " ".join([word for word in str(txt).split() if word not in STOPWORDS])
df['reviews_list']=df['reviews_list'].apply(lambda txt:remove_stopwords(txt))
df[['reviews_list']].sample(3)
```

```python
def remove_urls(txt):
    url_pattern=re.compile(r'https?://\S+/www\.\S+')
    return url_pattern.sub(r'',txt)
df['reviews_list']=df['reviews_list'].apply(lambda txt:remove_urls(txt))
df[['reviews_list']].sample(4)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def get_top_words(column,top_no_of_words,no_of_word):
    vec=CountVectorizer(ngram_range=no_of_word,stop_words='english')
    bag_of_words=vec.fit_transform(column)
    sum_words=bag_of_words.sum(axis=0)
    word_freq=[(word,sum_words[0,idx]) for word,idx in vec.vocabulary_.items()]
    word_freq=sorted(word_freq,key=lambda x:x[1],reverse=True)
    return word_freq[:top_no_of_words]
```

```python
# samples 60% of your dataframe
df_percent=df.sample(frac=0.6)
```

```python
df_percent.set_index('name',inplace=True)
```

```python
index=pd.Series(df_percent.index)
```

```python
tfidf=TfidfVectorizer(analyzer='word',ngram_range=(1,2),min_df=1,stop_words='english')
tfidf_matrix=tfidf.fit_transform(df_percent['reviews_list'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
def recommend(name,cosine_similarities=cosine_similarities):
    # create a list to put top restaurants
    recommend_restaurant=[]
    idx=index[index==name].index[0]
    # find the restaurants with a similar cosine similarity value
    score_series=pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
    # extract top 30 restaurant indexes with a similar cosine sim value
    top_30_indexes=list(score_series.iloc[0:31].index)
    # Names of the top 30 restaurants
    for each in top_30_indexes:
        recommend_restaurant.append(list(df_percent.index)[each])
    # create a new dataset
    df_new=pd.DataFrame(columns=['cuisines','Mean rating','cost'])
    for each in recommend_restaurant:
        df_new=pd.concat([df_new,df_percent[['cuisines','Mean rating','cost']].loc[[each]]],ignore_index=True)
        # drop the same named restaurants and sort only the top 10 by highest rating
        df_new=df_new.drop_duplicates(subset=['cuisines','Mean rating','cost'],keep=False)
        df_new=df_new.sort_values(by='Mean rating',ascending=False).head(10)
        print('TOP %s RESTAURANTS LIKE %s WITH SIMILAR REVIEWS:' % (str(len(df_new)),name))
        return df_new
recommend('Addhuri Udupi Bhojana')
```
